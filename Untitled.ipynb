```python
import pandas as pd
```

```python
#2. Load the dataset into a pandas DataFrame. 
df = pd.read_excel("_Social Media Usage Survey.xlsx")
df
```

```python
# quick count of missing values per column
df.isnull().sum() 
```

```python
# remove columns that are fully empty
df.dropna(axis=1, how="all", inplace=True) 
```

```python
df.shape
```

```python
# fill any missing cell with 'Unknown' text
df.fillna("Unknown", inplace=True)
```

```python
#Convert numeric column type
df["Daily Screen Time (hours)"] = pd.to_numeric(df["Daily Screen Time (hours)"], errors="coerce")
```
